```python
import numpy as np
import pandas as pd

np.random.seed(42)

# Generate random data for each feature
data = {
    'air_quality': np.random.rand(5000),
    'temperature': np.random.uniform(14, 25, 5000),
    'humidity': np.random.uniform(30, 60, 5000),
    'lighting': np.random.uniform(0, 1000, 5000),
    'noise_level': np.random.uniform(30, 70, 5000),
    'green_spaces': np.random.uniform(0, 1, 5000),
    'object_movements': np.random.randint(0, 10, 5000),
    'pollution_controls': np.random.choice(['Yes', 'No'], 5000),
    'elderly_health': np.random.choice(['Excellent', 'Good', 'Fair', 'Poor'], 5000),
    'wellbeing': np.random.choice(['High', 'Medium', 'Low'], 5000),
    'building_type': np.random.choice(['House', 'Apartment', 'Office'], 5000),
    'floor_level': np.random.choice(['Ground', '1st', '2nd', '3rd', '4th'], 5000),
    'window_type': np.random.choice(['Single', 'Double', 'Triple'], 5000),
    'ventilation_type': np.random.choice(['Natural', 'Mechanical'], 5000),
    'location': np.random.choice(['Urban', 'Suburban', 'Rural'], 5000)
}

# Create the DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the dataset
print(df.head())
```

```python
# Generate target feature based on criteria
df['healthy_environment'] = (
    (df['air_quality'] >= 0.5) &
    (df['temperature'] >= 18) &
    (df['humidity'] >= 40) &
    (df['lighting'] >= 250) &
    (df['noise_level'] <= 50) &
    (df['green_spaces'] >= 0.4)
).astype(int)

# Display the updated dataset
print(df.head())
```

```python
import numpy as np
from scipy import stats

# Generate a sample data following a Pareto distribution
pareto_data = np.random.pareto(2, size=1000)

# Apply the Box-Cox transformation to approximate a normal distribution
transformed_data, lambda_ = stats.boxcox(pareto_data)

# Plot the original and transformed data
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(pareto_data, bins=30)
axes[0].set_title('Original Data (Pareto Distribution)')

axes[1].hist(transformed_data, bins=30)
axes[1].set_title('Transformed Data (Approximating Normal Distribution)')

plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.stats import pareto, lognorm, weibull_min, gamma

np.random.seed(42)

# Generate random data for each feature
data = {
    'air_quality': np.random.rand(5000),
    'temperature': np.random.normal(20, 2, 5000),
    'humidity': pareto.rvs(2, size=5000),
    'lighting': np.random.lognormal(6, 1, 5000),
    'noise_level': weibull_min.rvs(1.5, size=5000),
    'green_spaces': np.random.normal(2, 5, 5000),
    'object_movements': np.random.gamma(2, 1, 5000),
    'pollution_controls': np.random.choice(['Yes', 'No'], 5000),
    'elderly_health': np.random.choice(['Excellent', 'Good', 'Fair', 'Poor'], 5000),
    'wellbeing': np.random.choice(['High', 'Medium', 'Low'], 5000),
    'building_type': np.random.choice(['House', 'Apartment', 'Office'], 5000),
    'floor_level': np.random.choice(['Ground', '1st', '2nd', '3rd', '4th'], 5000),
    'window_type': np.random.choice(['Single', 'Double', 'Triple'], 5000),
    'ventilation_type': np.random.choice(['Natural', 'Mechanical'], 5000),
    'location': np.random.choice(['Urban', 'Suburban', 'Rural'], 5000)
}

# Create the DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the dataset
df.head()
```

```python
# Generate the target feature based on conditions
df['healthy_environment'] = 0  # Initialize all values to 0

# Set values to 1 based on certain conditions
conditions = (
    (df['air_quality'] > 0.5) &
    (df['temperature'] > 18) &
    (df['humidity'] < 50) &
    (df['noise_level'] < 60) &
    (df['pollution_controls'] == 'Yes') &
    (df['wellbeing'] == 'High')
)
df.loc[conditions, 'healthy_environment'] = 1
```

```python
df['healthy_environment'].value_counts()
```

```python
import seaborn as sns
# Univariate Analysis
def univariate_analysis(data):
    # Plot histograms for numerical features
    numerical_features = ['air_quality', 'temperature', 'humidity', 'lighting', 'noise_level', 'green_spaces','object_movements']
    for feature in numerical_features:
        plt.figure(figsize=(8, 6))
        sns.histplot(data=data, x=feature, kde=True)
        plt.title(f'{feature} Distribution')
        plt.show()
    
    # Plot bar plots for categorical features
    categorical_features = [ 'pollution_controls', 'elderly_health', 'wellbeing',
                            'building_type', 'floor_level', 'window_type', 'ventilation_type', 'location']
    for feature in categorical_features:
        plt.figure(figsize=(8, 6))
        sns.countplot(data=data, x=feature)
        plt.title(f'{feature} Distribution')
        plt.xticks(rotation=45)
        plt.show()


# Perform univariate analysis
univariate_analysis(df)
```

```python
import numpy as np

# Define the number of rows to add null values to
num_rows_with_null = 98
# Select random row indices to add null values
random_row_indices = np.random.choice(df.index, num_rows_with_null, replace=False)
# Set null values for the selected row indices and specific columns
df.loc[random_row_indices, 'temperature'] = np.nan

# Define the number of rows to add null values to
num_rows_with_null = 124
# Select random row indices to add null values
random_row_indices = np.random.choice(df.index, num_rows_with_null, replace=False)
df.loc[random_row_indices, 'building_type'] = np.nan

# Define the number of rows to add null values to
num_rows_with_null = 164
# Select random row indices to add null values
random_row_indices = np.random.choice(df.index, num_rows_with_null, replace=False)
df.loc[random_row_indices, 'window_type'] = np.nan


```

```python
import numpy as np

# Define the number of outliers to add
num_outliers = 6

# Generate random outlier values
outliers = np.random.uniform(1.5, 2.0, num_outliers)

# Select random row indices to add outliers
random_row_indices = np.random.choice(df.index, num_outliers, replace=False)

# Set outliers for the selected row indices in the 'green_spaces' column
df.loc[random_row_indices, 'green_spaces'] = outliers

```

```python

# Outlier Analysis
def outlier_analysis(data):
    # Plot box plots for numerical features to identify outliers
    numerical_features = ['air_quality', 'temperature', 'humidity', 'lighting', 'noise_level', 'green_spaces']
    for feature in numerical_features:
        plt.figure(figsize=(8, 6))
        sns.boxplot(data=data, y=feature)
        plt.title(f'{feature} Outliers')
        plt.show()


# Perform outlier analysis
outlier_analysis(df)
```
